```python
xtree = et.parse("C:\\Users\\esrijyo\\Desktop\\batchMetadata.xml")
xroot = xtree.getroot()
```

```python
for node in xroot:
    #r_object_id = node.attrib.get("r_object_id")
    r_object_id = node.find("r_object_id").text
    i_chronicle_id = node.find("i_chronicle_id").text
    r_object_type = node.find("r_object_type").text
    is_current = node.find("is_current").text
    print(F" { r_object_id } { i_chronicle_id } { r_object_type } { is_current }")
    
    '''    
<r_object_id>09004d06822df9d9</r_object_id>
<i_chronicle_id>09004d06822df95b</i_chronicle_id>
<r_object_type>lpd_claim_chart</r_object_type>
<is_current>F</is_current>
'''
```

```python
import pandas as pd 

df_cols = ["r_object_id", "i_chronicle_id", "r_object_type","is_current"]
rows = []

for node in xroot:
    #r_object_id = node.attrib.get("r_object_id")
    r_object_id = node.find("r_object_id").text
    i_chronicle_id = node.find("i_chronicle_id").text
    r_object_type = node.find("r_object_type").text
    is_current = node.find("is_current").text

    rows.append({"r_object_id": r_object_id, "i_chronicle_id": i_chronicle_id, 
                 "r_object_type": r_object_type, "is_current": is_current})


out_df = pd.DataFrame(rows, columns = df_cols)
```
